```
from six.moves import urllib
import random
```

```
def get_images(file_name,num_train,num_test,output_path):
  
  f = open(file_name)
  url_list = [i for i in f]
  random.shuffle(url_list)
  f.close()
  
  img_iter = iter(url_list)
  control = 0
  while control < num_train:
    line = next(img_iter)
    id_img, img_url = line.split('\t')
    img_url = img_url.replace('\n','')
    img_format = img_url.split('.')[-1] 
    if img_format == 'jpg':
      name = output_path + 'train/' + id_img + '.jpg'
      try:
        urllib.request.urlretrieve(img_url,name)
        control += 1
        print(control)
      except:
        continue
    
  control = 0
  while control < num_test:
    line = next(img_iter)
    id_img, img_url = line.split('\t')
    img_url = img_url.replace('\n','')
    img_format = img_url.split('.')[-1]
    if img_format == 'jpg':
      name = output_path + 'test/' + id_img + '.jpg'
      try:
        urllib.request.urlretrieve(img_url,name)
        control += 1
        print(control)
      except:
        continue
```

```
!mkdir mini_imagenet
!mkdir mini_imagenet/train
!mkdir mini_imagenet/test
```

```
file_url = 'fall11_urls.txt' 
output_path = './mini_imagenet/'
num_train = 4000
num_test = 500
```

```
get_images(file_url,num_train,num_test,output_path)
```
